```python
age = input("Are you a cigarette addict older than 75 years old? Yes or No: ").strip().lower()
chronic = input("Do you have a severe chronic disease? Yes or No: ").strip().lower()
immune = input("Is your immune system too weak? Yes or No: ").strip().lower()

if age == "no":
    age = not bool(age)
if chronic == "no":
    chronic = not bool(chronic)
if immune == "no":
    immune = not bool(immune)
    
risk = age or chronic or immune

if risk:
    print("You are in risky group")
else:
    print("You are not in risky group")
```
